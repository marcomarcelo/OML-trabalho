```python
import requests
import pandas as pd
```

```python
# Ler os dados
file_path = '../../data/lending_data.csv'
df = pd.read_csv(file_path)
```

```python
# Linhas aleatórias do dataframe apenas para ter um exemplo de input
input_df = df.sample(10)
```

```python
input_df
```

```python
# Retiramos o target (default.payment.next.month) dos dados, para termos o input que o nosso modelo está à espera 
input_data = input_df.drop("default.payment.next.month", axis=1)
input_data
```

# FastAPI

```python
# Input de dados esperado
records = input_data.to_dict(orient="records")
request_dict = records[0]
request_dict
```

```python
# Fazer pedido à app com os inputs
response = requests.post("http://127.0.0.1:5002/default_payment", json=request_dict)
```

```python
# recebemos uma resposta com a previsão
response.json()
```
